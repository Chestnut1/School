```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

image = 255 * np.ones((400,300,3), np.uint8)

image[200:400,0:150] = [255,0,0]
image[0:200,0:300] = [0,255,0]
image[0:200,150:300] = [0,0,255]

masky = np.random.normal(75,20,10000).round().astype(np.int)
maskx = np.random.normal(100,35,10000).round().astype(np.int)

image[maskx,masky] = [0,0,0]

masky = np.random.normal(225,16,10000).round().astype(np.int)
maskx = np.random.normal(300,20,10000).round().astype(np.int)

maskx[maskx >= 400] -= 200
masky[masky >= 300] -= 150

image[-maskx,-masky] = [0,0,0]

plt.imshow(image)
```
