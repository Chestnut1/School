```python
import numpy as np

list = [1,2,3]
arr = np.array(list)

print("list +", list + list) # l'operatore + coincide con la concatenazione       
print("array +", arr + arr)  # l'operatore + Ã¨ applicato elemento per elemento: ufunc    
                             # le operazioni sono fatte elemento per elemento su numpy 
    
print("list *", list * 2)    # raddoppia la lista
print("array *", arr * 2)    # raddoppia ogni elemento dell'array


#se voglio concatenare su np faccio cosi':

a1 = np.array([[0, 1], [2, 3]])
a2 = np.array([[4, 5]])

print("axis=0 \n", str(np.concatenate((a1, a2), axis=0)))              # di default
print("axis=1 \n", str(np.concatenate((a1, a2.transpose()), axis=1)))
print("axis=None\n", str(np.concatenate((a1, a2), axis=None)))         # flatten


#per impilarle
a1 = np.array([0, 1])
a2 = np.array([2, 3])

print("axis=0 \n", np.stack((a1, a2)))            # sottinteso axis=0
print("axis=1 \n", np.stack((a1, a2), axis=-1))   #inverte riga con colonna 

#come faccio su python a fare il quadrato di ciascun elemento
list = [1,2,3]

list = [a ^ 2 for a in list]

#in np invece 
print(arr**2)

#in np anche il simbolo di ugualianza mette a confronto due liste,
#restituendo un vettore di booleani con True dove gli elementi combaciano,
#e False dove no

a = np.array([1,2,3,4])
b = np.array([1,2,3,5])

#dove sono ufuali metto -1
a[a == b] = -1
print(a)


#per arrotondare

a = np.array([-1.51, -0.22, 0., 0.27, 1.56])
print(a)
print("rint: ",np.rint(a))    #arrotonda al primo intero -1.51 diventa -2
print("floor:",np.floor(a))   #approssima per difettp
print("ceil: ",np.ceil(a))    #approssima per eccesso
print("trunc:",np.trunc(a))   #tronca
print("around",np.around(a,1))# arrotonda al decimale specificato. in questo caso 1



#altra roba

a = np.array([[1,2],[3,4]])

print(a)

print("sum(): ", a.sum())                     #somma tutti gli elementi del vettore
print("prod(axis=1):",a.prod(axis=1))         #fa il prodotto per gli assi. Se metto 0 mi da [3, 8]
                                              #senza axis mi prende l'array come flatten
print("min(axis=0):", a.min(axis=0))
print("max(axis=1):", a.max(axis=1))
print("argmin(axis=0):", a.argmin(axis=1))
print("argmax():", a.argmax())








```
